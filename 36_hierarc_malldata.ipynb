```python
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/datasets/Mall_Customers.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Genre']=le.fit_transform(df['Genre'])
```

```python
from sklearn.preprocessing import normalize
data=normalize(df)
data
```

```python
df1=pd.DataFrame(data,columns=df.columns)
df1
```

```python
import scipy.cluster.hierarchy as shc                #scientific python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.title('DENDROGRAM')
den=shc.dendrogram(shc.linkage(df1,method='average'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='average')
cluster.fit_predict(df1)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(df1['Annual Income (k$)'],df1['Spending Score (1-100)'],c=cluster.labels_)
```
